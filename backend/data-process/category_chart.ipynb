```python
all_cat_df = pd.merge(categories_df, expenses_df, left_on='id', right_on='category_id', how='left').groupby('name')['amount'].sum().fillna(0).reset_index()
all_cat_df.loc[all_cat_df['name'] == 'Ingresos', 'amount'] = incomes_df['amount'].sum()

all_cat_df.groupby(['name'])['amount'].sum().fillna(0).reset_index()
```

```python
expenses_subcat_df = pd.merge(subcategories_df, expenses_df, left_on='id', right_on='subcategory_id', how='left')
incomes_subcat_df = pd.merge(subcategories_df, incomes_df, left_on='id', right_on='subcategory_id', how='left')
```

```python
combined_test_df = expenses_subcat_df.merge(incomes_subcat_df, on='name', how='outer', suffixes=('_expense', '_income'))
```

```python
combined_test_df['amount_expense'].fillna(0, inplace=True)
combined_test_df['amount_income'].fillna(0, inplace=True)

combined_test_df['value'] = combined_test_df['amount_income'] + combined_test_df['amount_expense']
```

```python
def get_category_name(row):
    category_id = row['category_id_x']
    if pd.notna(category_id) and category_id in combined_test_df.index:
        return combined_test_df.loc[category_id, 'name']

    return None
```

```python
combined_test_df['category'] = combined_test_df.apply(get_category_name, axis=1)
```
