# Filter total account

```python
unique_accounts = combined_df['account_id'].unique()
dates_by_account = {}
amounts_by_account = {}
```

```python
unique_accounts
```

```python
for account in unique_accounts:
    # Filtrar el DataFrame por la cuenta actual
    account_df = combined_df[combined_df['account_id'] == account]

    # Agrupar por fecha y calcular la suma de montos para esa cuenta
    grouped = account_df.groupby('date')['amount'].sum().reset_index()

    # Extraer las fechas y montos como listas
    dates = grouped['date'].to_list()
    amounts = grouped['amount'].to_list()

    # Almacenar las listas en los diccionarios por cuenta
    dates_by_account[account] = dates
    amounts_by_account[account] = amounts
```
